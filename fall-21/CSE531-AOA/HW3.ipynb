```python
def get_max_partitions(arr):
    l = len(arr)
    num_p = [0] * (l+1)
    last_sum = [0] * (l+1)
    last_i = [0] * (l+1)
    
    num_p[1] = 1
    last_sum[1] = arr[0]
    last_i[1] = 0
    
    for i in range(2, l+1):
        print(f'\n\ni = {i}')
        max_idx = 0
        max_ps = 0
        max_sum = 0
        
        for j in range(1, i+1):
            print(f'\nj = {j}')
            print(f'max_idx = {max_idx}')
            print(f'max_ps = {max_ps}')
            print(f'max_sum = {max_sum}')
        
            s = 0    

            for k in range(i-j, i):
                print(f'k == i-j = {k}, arr[k] = {arr[k]}')
                s += arr[k]
                print(f'adding arr[k] to s; s = {s}')
            
            print(f'last_sum[i-j] = {last_sum[i-j]}')
            print(f'check s >= last_sum[i-j]')
            if s <= last_sum[i-j]:
                print(f'max_ps = {max_ps}, num_p[i-j] = {num_p[i-j]}')
                print(f'check max_ps < num_p[i-j] + 1')
                if max_ps > num_p[i-j] + 1:
                    max_ps = num_p[i-j] + 1
                    max_sum = s
                    max_idx = i-j
                    
        print(f'\nnum_p = {num_p}')
        print(f'last_sum = {last_sum}')
        print(f'last_i = {last_i}')
        num_p[i] = max_ps
        last_sum[i] = max_sum
        last_i[i] = max_idx
        
    i = l
    while i > 0:
        print(last_sum[i])
        i = last_i[i]
        
    return num_p, last_i, last_sum
```

```python
def max_part(arr, idx, prev_sum):
    if idx == len(arr):
        return 0
    s = 0
    ans = 0
    
    for i in range(idx, len(arr)):
        s += arr[i]
        temp = 0
        if s <= prev_sum:
            temp = 1 + max_part(arr, i+1, s)
        ans = max(temp, ans)
        
    return ans
```

```python
arr = [7, 5, 3, 5]
max_part(arr, 0, 9999)
```
