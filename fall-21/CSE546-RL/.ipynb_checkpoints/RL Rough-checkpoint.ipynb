```python
import numpy as np
```

```python
def right(gamma):
    return 0.5*gamma + 3*(gamma**3)
```

```python
def down(gamma):
    return 1.5*(gamma**2) + 3*(gamma**3)
```

```python
def diag(gamma):
    return -0.4 + 3*gamma
```

```python
for i in np.linspace(0, 1, 1000):
    res = ['right', 'down', 'diag']
    x = np.argmax(np.array([right(i), down(i), diag(i)]))
    print(i, res[x])
    
```
