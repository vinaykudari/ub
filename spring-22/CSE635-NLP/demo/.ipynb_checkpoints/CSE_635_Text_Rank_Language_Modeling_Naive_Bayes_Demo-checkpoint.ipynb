```
import numpy as np
import pandas as pd
import nltk
import re
from nltk.tokenize import sent_tokenize
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx
import copy
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```
df = pd.read_csv("tennis_articles.csv", encoding= 'unicode_escape')
df
```

```
df['article_text'][0]
```

```
# Combine all article_text and tokenizing into sentence

sentences = []
for s in df['article_text']:
  sentences.append(sent_tokenize(s))

sentences = [y for x in sentences for y in x]
```

```
len(sentences)
```

```
sentences[-10:]
```

```
# We will use Glove embeddings
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove*.zip
```

```
# Extract word vectors into a dictionary
word_embeddings = {}
f = open('glove.6B.100d.txt', encoding='utf-8')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    word_embeddings[word] = coefs
f.close()

len(word_embeddings)
```

```
# function to remove stopwords
def remove_stopwords(sen):
    sen_new = " ".join([i for i in sen if i not in stop_words])
    return sen_new
```

```
# remove punctuations, numbers, special characters, stopwords & Lowercase
clean_sentences = pd.Series(sentences).str.replace("[^a-zA-Z]", " ")
clean_sentences = [s.lower() for s in clean_sentences]
clean_sentences = [remove_stopwords(r.split()) for r in clean_sentences]
# clean_sentences = [i for i in clean_sentences if len(i) > 0]
```

```
sentences[:10]
```

```
clean_sentences[:10]
```

```
# For each sentence compute embedding by average pooling it's word embeddings
sentence_vectors = []
for i in clean_sentences:
  if len(i) != 0:
    v = sum([word_embeddings.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
  else:
    v = np.random.rand(100,)
  sentence_vectors.append(v)
```

```
len(sentence_vectors), len(clean_sentences)
```

```
# Iterative way of calculating similarity matrix
import time

st = time.time()
sim_mat = np.zeros([len(clean_sentences), len(clean_sentences)])

for i in range(len(clean_sentences)):
  for j in range(len(clean_sentences)):
    if i != j:
      sim_mat[i][j] = cosine_similarity(sentence_vectors[i].reshape(1,100), sentence_vectors[j].reshape(1,100))[0,0]

print(time.time()-st,"time taken!")
```

```
sim_mat.shape
```

```
sim_mat
```

```
# Pythonic way of calculating similarity matrix

st = time.time()
sentence_vector_arr = np.asarray(sentence_vectors)
sentence_vector_arr_norm = np.linalg.norm(sentence_vectors, axis=-1).reshape(sentence_vector_arr.shape[0], 1)

embedding_dot = np.dot(sentence_vector_arr, sentence_vector_arr.T)
norm_dot = np.dot(sentence_vector_arr_norm, sentence_vector_arr_norm.T)
sim_mat_2 = embedding_dot/norm_dot

similarity_mask = (np.eye(sim_mat_2.shape[0], sim_mat_2.shape[0]) == 0)*1.0
sim_mat_2 = sim_mat_2*similarity_mask
print(time.time()-st,"time taken!")
```

```
sim_mat_2
```

```
sim_mat_2.shape, sim_mat.shape
```

```
(sim_mat - sim_mat_2).sum()
```

```
# Compute Page Rank using Networkx
nx_graph = nx.from_numpy_array(sim_mat)
scores = nx.pagerank(nx_graph)
```

```
ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)

# Extract top 5 sentences as the summary
for i in range(5):
  print(ranked_sentences[i][1].strip(), "\n")
```

```
# Compute Page Rank using your own implementation
alpha = 0.15
N = sim_mat_2.shape[0]
P = ((sim_mat_2/sim_mat_2.sum(-1).reshape(-1,1)) * (1-alpha)) + alpha/N
```

```
P = np.nan_to_num(P, nan=1/N)
```

```
P.sum(-1)
```

```
A = np.eye(N)[0]
for i in range(40):
    A = A @ P
    # print(i, A)
```

# Simple Language Model

```
import nltk
nltk.download('reuters')
nltk.download('punkt')
from nltk.corpus import reuters
from nltk import bigrams, trigrams
from collections import Counter, defaultdict
from tqdm import tqdm
```

```
# Create a placeholder for model
model = defaultdict(lambda: defaultdict(lambda: 0))
```

```
sentences = reuters.sents()
print(len(sentences))
```

```
print(sentences[0])
```

```
" ".join(sentences[0])
```

```
list(trigrams(sentences[0], pad_right=True, pad_left=True))
```

```
# Count frequency of co-occurance  
for sentence in tqdm(sentences):
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model[(w1, w2)][w3] += 1
```

```
model[('trade', 'friction')]
```

```
# Let's transform the counts to probabilities
for w1_w2 in tqdm(model):
    total_count = float(sum(model[w1_w2].values()))
    for w3 in model[w1_w2]:
        model[w1_w2][w3] /= total_count
```

```
model[('trade', 'friction')]
```

```
import random

# starting words
text = ['today', 'the']
sentence_finished = False
 
while not sentence_finished:
  # select a random probability threshold  
  r = random.random()
  accumulator = .0

  for word in model[tuple(text[-2:])].keys():
      accumulator += model[tuple(text[-2:])][word]
      # select words that are above the probability threshold
      if accumulator >= r:
          text.append(word)
          break

  if text[-2:] == [None, None]:
      sentence_finished = True
 
print (' '.join([t for t in text if t]))
```

```
import numpy as np
import random

# starting words
text = ['today', 'the']
eos_tokens = [".","!","?"]
sentence_finished = False
 
while not sentence_finished:
  # select a random probability threshold  
  r = random.random()
  accumulator = .0
  available_next_word = dict(model[tuple(text[-2:])])
  next_word = np.random.choice(list(available_next_word.keys()), 1, p=list(available_next_word.values()))[0]
  text.append(next_word)
  for eos in eos_tokens:
    if next_word.endswith(eos):
      sentence_finished = True
      break
 
print (' '.join([t for t in text if t]))
```

# Text Classification using SKlearn

```
from sklearn.datasets import fetch_20newsgroups
from pprint import  pprint

newsgroups_train = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'))

print(newsgroups_train.filenames.shape, newsgroups_train.target.shape)
print(newsgroups_test.target.shape, newsgroups_test.target.shape)
```

```
{ix:i for ix, i in enumerate(list(newsgroups_train.target_names))}
```

```
pprint(list(zip(newsgroups_train.target, newsgroups_train.data))[:2])
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(newsgroups_train.data)
vectors.shape
```

```
np.asarray(vectors[0])
```

```
vectors_test = vectorizer.transform(newsgroups_test.data)
vectors_test.shape
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

clf = MultinomialNB(alpha=.01)
clf.fit(vectors, newsgroups_train.target)
pred = clf.predict(vectors_test)
```

```
print("Macro F1: ",metrics.f1_score(newsgroups_test.target, pred, average='macro'))
print("Micro F1: ",metrics.f1_score(newsgroups_test.target, pred, average='micro'))
print("Accuracy: ",metrics.accuracy_score(newsgroups_test.target, pred))
```

```
print(metrics.classification_report(newsgroups_test.target, pred))
```
