### Feature Engineering

```python
desc_gp = data.groupby('Violation Type').agg({'Description':lambda x:' '.join(x.values.astype(str))})
desc_gp.head()
```

```python
def get_counts(n, text):
    text = text.replace('\'', '')
    tokens = word_tokenize(text)
    text = Text(tokens)
    counts = {}
    
    for i in range(1, n+1):
        counts[i] = FreqDist(ngrams(text, i))
        
    return counts
```

```python
cit_c = get_counts(4, desc_gp.loc['Citation'].values[0])
sero_c = get_counts(4, desc_gp.loc['SERO'].values[0])
war_c = get_counts(4, desc_gp.loc['Warning'].values[0])
```

#### Using text classification on decription column to predict class and use that as a feature

```python
mp = {
    'Citation': 0,
    'SERO': 1,
    'Warning': 2,
}

rev_mp = {
    0: 'Citation',
    1: 'SERO',
    2: 'Warning',
}

X_train, X_test, y_train, y_test = train_test_split(data['Description'], data['Violation Type'], random_state = 0)
count_vect = CountVectorizer(max_features=1000, ngram_range=(1, 3), analyzer = 'word')
tfidf_transformer = TfidfTransformer()
X_train_counts = count_vect.fit_transform(X_train)
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
clf = MultinomialNB().fit(X_train_tfidf, y_train)

data['Description Pred'] = data['Description'].apply(lambda x: mp[clf.predict(count_vect.transform([x]))[0]])
```

```python
X_test_counts = count_vect.fit_transform(X_test)
X_test_tfidf = tfidf_transformer.fit_transform(X_test_counts)
clf.score(X_test_counts, y_test)
```

### Modelling

```python
target_col = 'Violation Type'
target = df[target_col]
select_cols = [
    'Contributed To Accident',
    'Description',
    # 'Driver Race',
    'Property Damage',
    'Personal Injury',
    'Description Pred',
]
X_train, x_test, y_train, y_test = train_test_split(df[select_cols], target, train_size=0.80)
```

#### Random Forest Classifier

```python
forest = RandomForestClassifier(n_jobs=-1, n_estimators=2000)
forest.fit(X_train, y_train)
forest_y = forest.predict(x_test)

test_score = forest.score(x_test, y_test)
print(f"Accuracy of the RandomForestClassifier: {test_score:.4f}")
```

```python
cv_results = cross_validate(forest, df[select_cols], target)
scores = cv_results["test_score"]
print(f"Accuracy score via cross-validation:\n"
      f"{scores.mean():.3f} +/- {scores.std():.3f}")
```

```python
print(classification_report(y_test, forest_y, target_names=['Citation', 'SERO', 'Warning']))
```

```python
res = []
for l in [0,1,2]:
    prec,recall,_,_ = precision_recall_fscore_support(
        np.array(y_test)==l,
        np.array(forest_y)==l,
        pos_label=True,
        average=None,
    )
    res.append([rev_mp[l], recall[0], recall[1]])

pd.DataFrame(res,columns = ['class','sensitivity','specificity'])
```

#### KNN

```python
knn = KNeighborsClassifier(300)
knn.fit(X_train, y_train)

knn_y = knn.predict(x_test)

test_score = knn.score(x_test, y_test)
print(f"Accuracy of the KNN: {test_score:.2f}")
```

```python
cv_results = cross_validate(knn, df[select_cols], target)
scores = cv_results["test_score"]
print(f"Accuracy score via cross-validation:\n"
      f"{scores.mean():.3f} +/- {scores.std():.3f}")
```

```python
print(classification_report(y_test, knn_y, target_names=['Citation', 'SERO', 'Warning']))
```

```python
res = []
for l in [0,1,2]:
    prec,recall,_,_ = precision_recall_fscore_support(
        np.array(y_test)==l,
        np.array(knn_y)==l,
        pos_label=True,
        average=None,
    )
    res.append([rev_mp[l], recall[0], recall[1]])

pd.DataFrame(res,columns = ['class','sensitivity','specificity'])
```
