### Inference

I have divided the data in 80:20 ratio for training and testing. After preprocessing and modelling I've found that decision tree gives the best performance compared to logistic regression and K nearest neighbours. With decision tree accuracy is around 87.5% with F measures of 0.89 and 0.82 respectively for survived and not survived classes. Also I have observed that the difference in test and train accuracy is least for decision tree model hence it is more generalizable and I will deploy decision tree model in production
